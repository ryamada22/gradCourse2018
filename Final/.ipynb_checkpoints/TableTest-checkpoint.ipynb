# 分轄表の検定

統計遺伝学でよく登場する2 by 3分割表の検定とハーディワインバーグ平衡検定を題材に、

Fishreの正確確率検定、カイ二乗検定、尤度比検定、コクラン・アーミテージ傾向性検定について、

理論、実施手順、Ｒコードをまとめる。分轄表検定やHWE検定などの離散的なデータの仮説検定には、

帰無仮説の下での観測データの生起確率を正確に計算してP値をえる正確検定と、

ある統計量がサンプルサイズを無限大に増やしていくことで特定の理論分布に漸近的に近づいていくことを用いてP値を計算する漸近検定の２種類がある。

上で挙げた中では、Fishreの正確確率検定は正確検定、カイ二乗検定、尤度比検定、傾向性検定は漸近検定である。

また、最後に分轄表の自由度と特徴統計量、可視化について2 by 3分轄表を例にまとめる。


## Fishreの正確確率検定

分割表のFishreの正確確率検定は観測されうる全ての分割表について、

行と列が独立であるという帰無仮説の下での生起確率を計算し、

観測された分割表とそれより低い生起確率を持つ分割表の生起確率の総和をp値として用いて

分割表の珍しさを検定する手法である。

STEP1:インプットデータとして分割表を用意

STEP2:周辺度数を固定化した場合の全ての分轄表の列挙

STEP3:各分割表の帰無仮説の下での生起確率を計算

STEP4:観測データのP値を計算

以下のＲコードでこれらを順に実行していく。

### STEP1 : データを用意

```r
#以下のようなcase/control by 3 genotypesの分割表を考える
tab <- matrix(c(9,42,49,25,50,25),2,3,byrow=T)
rownames(tab) <- c("case","control")
colnames(tab) <- c("AA","Aa","aa")
tab
```

## STEP2 : 分割表の列挙

周辺度数を固定した2 by 3 の分割表の自由度は2である。

そのため、以下のコードで全ての分割表を列挙できる。

```r
#STEP2 : 分割表の列挙
options(repr.plot.width=5, repr.plot.height=5)
cell11_can <- 0:min(sum(tab[1,]),sum(tab[,1]))
cell12_can <- 0:min(sum(tab[1,]),sum(tab[,2]))
cell_can <- expand.grid(cell11_can,cell12_can)
check_satis <- rep(NA,nrow(cell_can))
for(i in 1:nrow(cell_can)){
    tab_can <- matrix(NA,2,3)
    tab_can[1,1] <- cell_can[i,1]
    tab_can[1,2] <- cell_can[i,2]
    tab_can[1,3] <- sum(tab[1,]) - tab_can[1,1] - tab_can[1,2] #num of case is fixed
    tab_can[2,1] <- sum(tab[,1]) - tab_can[1,1] #num of AA is fixed
    tab_can[2,2] <- sum(tab[,2]) - tab_can[1,2] #num of Aa is fixed
    tab_can[2,3] <- sum(tab[,3]) - tab_can[1,3] #num of aa is fixed
    check_satis[i] <- all(tab_can>=0) & (sum(tab_can[2,])==sum(tab[2,]))
}
case_num <- sum(tab[1,])
cell_can_sub <- cell_can[check_satis,]
plot(cell_can_sub[,1],cell_can_sub[,2],
     xlab="table['case','AA']",ylab="table['case','Aa']")
```

### STEP3 生起確率を計算

周辺度数を固定したときの分割表の生起確率は以下の手順で計算できる。

これは、「２００人の中からcaseとして100人を選ぶ組み合わせのうち、

AAの人から9人、Aaの人から42人、aaの人から49人選ばれる確率はいくつか」という問題に言い換えられる。

例えば、上の分割表の場合、200人の中から100人のcaseを選ぶ組み合わせは200C100であり、これが分母になる。

上記のような分割表の場合の数は「AAの人34人から9人を選ぶ組み合わせ」　×　

「AAの人92人から42人を選ぶ組み合わせ」×「aaの人74人から49人を選ぶ組み合わせ」になるため、

生起確率は以下のように表せる.

$$
\frac{{}_{34} C _9 * {}_{92} C _{42} * {}_{74} C _{49}}{{}_{200} C _{100}}
$$


```r
#上記の式に基づいて周辺度数を固定した場合の全ての分割表の生起確率を計算する
n <- sum(tab)
case_num = sum(tab[1,])
cont_num = sum(tab[2,])
AA_num = sum(tab[,1])
Aa_num = sum(tab[,2])
aa_num <- sum(tab[,3])
ocprob <- rep(NA,nrow(cell_can_sub))
for(i in 1:nrow(cell_can_sub)){
    tab_can <- matrix(NA,2,3)
    tab_can[1,1] <- cell_can_sub[i,1]
    tab_can[1,2] <- cell_can_sub[i,2]
    tab_can[1,3] <- case_num - tab_can[1,1] - tab_can[1,2] #num of case is fixed
    tab_can[2,1] <- AA_num - tab_can[1,1] #num of AA is fixed
    tab_can[2,2] <- Aa_num - tab_can[1,2] #num of Aa is fixed
    tab_can[2,3] <- aa_num - tab_can[1,3] #num of aa is fixed
    tmp <- choose(AA_num,tab_can[2,1]) * choose(Aa_num,tab_can[2,2]) * choose(aa_num,tab_can[2,3])
    ocprob[i] <- tmp/choose(n,case_num)
}

#カラーパレット作成
color_norm1 <- cm.colors(length(ocprob))
color_norm1 <- color_norm1[rank(ocprob)]
plot(cell_can_sub[,1],cell_can_sub[,2],col=color_norm1,xlab="table['case','AA']",ylab="table['case','Aa']")
```

```r
#生起確率の総和が１になっていることの確認
sum(ocprob)
```

### STEP4 P値を計算

P値は、「自分以下の生起確率をもつ分割表が観測される確率」として計算できるので、

生起確率の一覧からP値を計算でき、これが正確確率検定によるP値になる。

分割表のうちP<0.01になるものを赤でプロットする. 

中央部分のだ円形の領域はありふれた分割表であり、その外側の分割表はP値が低い、つまり珍しい分割表になる。

```r
#P値の描画
pval <- rep(NA,length(ocprob))
for(i in 1:length(ocprob)){
    pval[i] <- sum(ocprob[ocprob<=ocprob[i]])
}
cols <- ifelse(pval<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']")
```

## 分割表のカイ２乗検定

分割表のカイ二乗検定は以下のステップで行なう。

STEP1:帰無仮説の下で得られる各セルの実現確率の理論値を設定する。

STEP2：理論値の下で計算される各セルの個数の期待値と観測値から計算される以下の統計量を算出する。

$$
\sum\frac{(x_{obs} - x_{exp})^2}{x_{exp}}
$$

サンプル数を増やすと、帰無仮説が真の場合には漸近的にカイ二乗分布に近づくことを用いてP値を算出する。

そのため、「2 by 3の分割表」において、「列と行が独立である」という帰無仮説を検定する場合、

各セルの期待カウント数は、周辺度数を用いて計算できる。

たとえば、case && AAのセルの期待度数は$\frac{AA*case}{N}$となり、カイ二乗分布の自由度は２である。

N by M の分割表の場合でも同様に計算でき、用いるカイ二乗分布の自由度は(N-1)(M-1)になる。

```r
## カイ二乗検定を行う関数
mychisq <- function(tab){
    tab_exp <- matrix(NA,2,3)
    N <- sum(tab)
    case <- sum(tab[1,])
    control <- sum(tab[2,])
    AA <- sum(tab[,1])
    Aa <- sum(tab[,2])
    aa <- sum(tab[,3])
    tab_exp[1,1] <- (case*AA)/N
    tab_exp[1,2] <- (case*Aa)/N
    tab_exp[1,3] <- (case*aa)/N
    tab_exp[2,1] <- (control*AA)/N
    tab_exp[2,2] <- (control*Aa)/N
    tab_exp[2,3] <- (control*aa)/N
    chisq <- sum(((tab - tab_exp)^2)/tab_exp)
    pvalue <- pchisq(chisq,df=2,lower.tail = F)
    return(pvalue)
}
```

```r
#上記の式に基づいて周辺度数を固定した場合の全ての分割表の生起確率を計算する
n <- sum(tab)
case_num = sum(tab[1,])
cont_num = sum(tab[2,])
AA_num = sum(tab[,1])
Aa_num = sum(tab[,2])
aa_num <- sum(tab[,3])
pval_chisq <- rep(NA,nrow(cell_can_sub))
for(i in 1:nrow(cell_can_sub)){
    tab_can <- matrix(NA,2,3)
    tab_can[1,1] <- cell_can_sub[i,1]
    tab_can[1,2] <- cell_can_sub[i,2]
    tab_can[1,3] <- case_num - tab_can[1,1] - tab_can[1,2] #num of case is fixed
    tab_can[2,1] <- AA_num - tab_can[1,1] #num of AA is fixed
    tab_can[2,2] <- Aa_num - tab_can[1,2] #num of Aa is fixed
    tab_can[2,3] <- aa_num - tab_can[1,3] #num of aa is fixed
    pval_chisq[i] <- mychisq(tab_can)
}
cols <- ifelse(pval_chisq<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']")
```

## 尤度比検定

2 by 3分割表の尤度比検定は以下のステップで行なう。

尤度比検定は、「帰無仮説のもとで観測データが得られる尤度の最大値」と

「対立仮説の下で観測データが得られる尤度の最大値」の比をもとめ、これを2log()した統計量が

サンプル数を増やすと帰無仮説が真の場合には漸近的に自由度2のカイ二乗分布に近づくことを用いてP値を算出する。

「帰無仮説のもとで観測データが得られる尤度の最大値」は、カイ２乗検定のときと同様の期待分割表の頻度から計算する。

「対立仮説の下で観測データが得られる尤度は」、分轄表の各セルの存在割合がそのまま各セルの確率の最尤推定値になる。

このことを用いると、尤度比$L_{alt}/L_{null}$は$\prod{tab_{i,j}/tabexp_{i,j}}$と表せる。

N by M の分割表の場合でも同様に計算でき、用いるカイ二乗分布の自由度は(N-1)(M-1)になる。

```r
#期待分割表を返す関数
makeexptab <- function(tab){
    m1 <- apply(tab,1,sum)
    m2 <- apply(tab,2,sum)
    N <- sum(tab)
    etab <- m1 %*% t(m2)/N
    return(etab)
}

## 尤度比検定を行う関数
mylr <- function(tab){
    tab_exp <- makeexptab(tab)
    chisq <- 2 * sum(tab * log(tab/tab_exp))
    pvalue <- pchisq(chisq,df=2,lower.tail = F)
    return(pvalue)
}

mylr(tab)
```

```r
#上記の式に基づいて周辺度数を固定した場合の全ての分割表の生起確率を計算する
n <- sum(tab)
case_num = sum(tab[1,])
cont_num = sum(tab[2,])
AA_num = sum(tab[,1])
Aa_num = sum(tab[,2])
aa_num <- sum(tab[,3])
pval_chisq <- rep(NA,nrow(cell_can_sub))
for(i in 1:nrow(cell_can_sub)){
    tab_can <- matrix(NA,2,3)
    tab_can[1,1] <- cell_can_sub[i,1]
    tab_can[1,2] <- cell_can_sub[i,2]
    tab_can[1,3] <- case_num - tab_can[1,1] - tab_can[1,2] #num of case is fixed
    tab_can[2,1] <- AA_num - tab_can[1,1] #num of AA is fixed
    tab_can[2,2] <- Aa_num - tab_can[1,2] #num of Aa is fixed
    tab_can[2,3] <- aa_num - tab_can[1,3] #num of aa is fixed
    pval_chisq[i] <- mylr(tab_can)
}
cols <- ifelse(pval_chisq<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']")
```

## コクラン・アーミテージ傾向性検定

コクラン・アーミテージ傾向性検定は、2 by 3分割表において列に順序がある場合、

各列ごとのcaseの割合と順序に対して直線的な増加・減少の関係があるかどうかを検定する。
（https://www.weblio.jp/content/コクラン・アーミテージ検定）.

そのため、3 genotypeにどのような順序の入れ方をするかによってAdditiveモデルやDiminanceモデルなどいろいろな

SNPと形質の関連パターンを表現することができる。順序の入れ方によって、どのような仮説を検定するかが異なるため、

どのような分割表でP値が低くなるかが異なる。

```r
#Additive model, Dominance model, Recessive modelでのP<0.01となる分割表の領域の描画
res_p_add <- rep(NA,nrow(cell_can_sub))
res_p_dom <- rep(NA,nrow(cell_can_sub))
res_p_rec <- rep(NA,nrow(cell_can_sub))
for(i in 1:nrow(cell_can_sub)){
    tab_can <- matrix(NA,2,3)
    tab_can[1,1] <- cell_can_sub[i,1]
    tab_can[1,2] <- cell_can_sub[i,2]
    tab_can[1,3] <- sum(tab[1,]) - tab_can[1,1] - tab_can[1,2] #num of case is fixed
    tab_can[2,1] <- sum(tab[,1]) - tab_can[1,1] #num of AA is fixed
    tab_can[2,2] <- sum(tab[,2]) - tab_can[1,2] #num of Aa is fixed
    tab_can[2,3] <- sum(tab[,3]) - tab_can[1,3] #num of aa is fixed
    res_p_add[i] <- prop.trend.test(tab_can[1,],colSums(tab_can),c(0,1,2))$p.value
    res_p_dom[i] <- prop.trend.test(tab_can[1,],colSums(tab_can),c(0,1,1))$p.value
    res_p_rec[i] <- prop.trend.test(tab_can[1,],colSums(tab_can),c(0,0,1))$p.value
}
```

```r
par(mfrow=c(2,2))
cols <- ifelse(res_p_add<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']",main="Additive")
cols <- ifelse(res_p_dom<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']",main="Dominant")
cols <- ifelse(res_p_rec<0.01,"red","black")
plot(cell_can_sub[,1],cell_can_sub[,2],col=cols,xlab="table['case','AA']",ylab="table['case','Aa']",,main="recessive")
```

## ハーディワインバーグ平衡の正確確率検定・カイ二乗検定・尤度比検定

HWE検定は、ある集団の3 genotype(AA,Aa,aa)を持つサンプルの数が観測されたとき、

これは上の考え方を使うことで正確確率検定、カイ二乗検定、尤度比検定を行なうことができる.

### HWEの正確確率検定

正確確率検定を行なうためには、「分割表の行と列が独立であるか」という命題に問題を変換する必要がある。

これは、行に各個人、列に2列(A,a)の分割表を作り、集団中の各個人がA,aアリルをそれぞれいくつずつ持っているか

という分割表に変換できる。例えば、AA,Aa,aaが２、４、２人の場合、これを分割表であらわすと以下のようになる。

```r
n_2 <- 2
n_1 <- 4
n_0 <- 2
N <- n_2 + n_1 + n_0
mat <- rbind(cbind(rep(2,n_2),rep(0,n_2)),cbind(rep(1,n_1),rep(1,n_1)),cbind(rep(0,n_0),rep(2,n_0)))
rownames(mat) <- paste0("sample",1:N)
colnames(mat) <- c("A","a")
mat
```

```r
geno <- c(2,6,2)
names(geno) <- c("AA","Aa","aa")
```

この分割表の生起確率はN by M分割表の生起確率の式より、

$$
\frac{{_2C_2}^{n_2} * {_2C_1}^{n_1} * {_2C_0}^{n_0}}{_{2N}C_{n_A}}
$$

ここで、n0人、n1人、n2人の行を区別はないので、

各genotype人数の生起確率probは、

$$
prob = \frac{{_2C_2}^{n_2} * {_2C_1}^{n_1} * {_2C_0}^{n_0}}{_{2N}C_{n_A}} * \frac{N!}{n_0! n_1! n_2!}
= \frac{n_A! n_a! 2^{n_1} N!}{(2N)! n_0! n_1! n_2!}
$$

このように計算される確率は非常に大きい値が登場するため計算のさい桁があふれてしまうことがある。

それを防ぐため、Rなどで計算する際はlog(prob）を計算し、最期にexponentialでもとに戻すという工夫ができる.

$$
log(prob) = log(n_A!) + log(n_a!) + n_1 log2 + log(N!) - log((2N)!) -  log(n_0!) - log(n_1!) - log(n_2!)
$$

log(n!)は、Rではlfactorial(n)で記述できる。

```r
#genotypeのカウント数のベクトルを入れると、その正確確率を返す関数
HWE_exact_prob <- function (geno){
  N <- sum(geno)
  n_2 <- geno[1]
  n_1 <- geno[2]
  n_0 <- geno[3]
  n_A <- 2*n_2 + n_1
  n_a <- 2*N - n_A
  log_prob <- lfactorial(n_A) + lfactorial(n_a) + n_1*log(2) + lfactorial(N) - lfactorial(2*N) -
              lfactorial(n_0) - lfactorial(n_1) - lfactorial(n_2)
  prob <- exp(log_prob)
  return(prob)
}

HWE_exact_prob(c(2,2,4))
```

上記の関数により各分割表の生起確率が計算できるので、2 by 3 分割表の場合と同様に、

$n_A$, $n_a$を固定したときに観測しうる全パターンのジェノタイプカウント数ベクトルを列挙し、

その生起確率分布を計算することで、観測されたジェノタイプカウントベクトルの正確確率検定のP値を算出できる。


```r
#
HWE_exact_test <- function(geno){
    N <- sum(geno)
    n_2 <- geno[1]
    n_1 <- geno[2]
    n_0 <- geno[3]
    n_A <- 2 * n_2 + n_1
    n_a <- 2*N - n_A
    max_geno1 <- ifelse(n_A%%2==0, n_A/2, (n_A-1)/2)
    #check_status <- rep(NA,max_geno1+1)
    ocprobs <- rep(NA,max_geno1+1)
    for(i in 1:length(ocprobs)){
        geno_tab <- rep(NA,3)
        geno_tab[1] <- i-1
        geno_tab[2] <- n_A - 2*geno_tab[1]
        geno_tab[3] <- N - geno_tab[1] - geno_tab[2]
        #cat(geno_tab,"\n")
        if(any(geno_tab<0)) cat("WARNING")
        ocprobs[i] <- HWE_exact_prob(geno_tab)
    }
    idx <- n_2 + 1
    p_value <- sum(ocprobs[ocprobs<=ocprobs[idx]])
    return(list(ocprobs,p_value))
}

result <- HWE_exact_test(c(20,20,40))
    
#総和は１になっている
sum(result[[1]])

#観測されたGenotypeのP値
result[[2]]
```

### HWEのカイ二乗検定

HWEもと期待される 3 genotypeの頻度を用いて同様の手順で検定を行なうことができる.

```r
HWE_chisq_test <- function(geno){
  N <- sum(geno)
  p <- (2*geno[1] + geno[2])/(2*N)
  q <- 1-p
  geno_exp <- N * c(p^2, 2*p*q, q^2)
  chisq <- sum((geno-geno_exp)^2/geno_exp)
  p <- pchisq(chisq,df=1,lower.tail = F)
  return(p)
}

#実行例
HWE_chisq_test(c(20,20,40))
```

### HWEの尤度比検定

HWEのもと期待される 3 genotypeの頻度を用いて同様の手順で検定を行なうことができる.

帰無仮説が成り立たない状況下でのもっとも尤度の高い3 genotypeの頻度は$\frac{AA}{n},\frac{Aa}{n},\frac{aa}{n}$

なので、これとHWEのもとでのgenotype頻度である$p^2, 2pq, q^2$の尤度を比較することになる.

```r
HWE_lr_test <- function(geno){
  N <- sum(geno)
  p <- (2*geno[1] + geno[2])/(2*N)
  q <- 1-p
  geno_exp <- c(p^2, 2*p*q, q^2)
  geno_alt <- c(geno[1]/N,geno[2]/N,geno[3]/N)
  chisq <- -2 * sum(geno*log(geno_exp/geno_alt))
  p <- pchisq(chisq,df=1,lower.tail = F)
  return(p)
}

HWE_lr_test(c(20,20,40))
```

## 分割表の自由度、特徴統計量、可視化

N by Mの分割表の自由度は、(N-1)(M-1)である。そのため、2 by 3の分割表では自由度２になる。

そのため、周辺度数を固定した2 by 3分割表は、各セル間が満たす独立な関係式２つで一意に特定できる。

このような分割表の特徴統計量としてよく用いられる指標として、リスク比、オッズ比がある。

例えば、以下の2 by 3 分割表を考える。

```r
tab <- matrix(c("a","b","c","d","e","f"),2,3,byrow=T)
rownames(tab) <- c("case","control")
colnames(tab) <- c("AA","Aa","aa")
tab
```

AA, Aa ,aa のジェノタイプの違いでどれだけ病気へのなりやすさがかわるかを定量したい。

AaのAAに対するリスク比は、$\frac{b}{b+e} / \frac{a}{a+d}$, 

aaのAAに対するリスク比は$\frac{c}{c+f} / \frac{a}{a+d}$

周辺度数が固定されているとき、これらを定めると分割表は一意に決まる。

一方で、オッズとはある事象の起きる確率と起きない確率の比であり、オッズ比は各水準同士でこの比をとったものである。

AA, Aa, aaのオッズはそれぞれ$\frac{a}{d}$, $\frac{b}{e}$, $\frac{c}{f}$である。

したがって、AAに対するAaのオッズ比は$\frac{bd}{ae}$, AAに対するaaのオッズ比は$\frac{cd}{af}$であり、

これらを定めても分轄表は一意に定まる。

また、自由度nの分割表は、n次元の正単体上の点として表現することができる。

例えば、2 by 3の分割表の場合には各分割表は正三角形の周上または内部の点として表現できる。

上では、x軸をcase && AAのサンプル数、y軸をcase && Aaのサンプル数として2次元プロットを描いたが、

ここで全ての点は６角形の領域におさまっているが、これは周辺度数を用いた以下のような6つの不等式で定まる領域である。

$$
0 < x \\
x < min(case,AA)\\
0 < y \\
y < min(case,Aa)\\
x + y < case \\
x + y < case - aa \\
$$

```r
#Visualization
plot(cell_can_sub[,1],cell_can_sub[,2],xlab="table['case','AA']",ylab="table['case','Aa']")
```

ここで座標軸を以下のような行列で線形変換すると、$0<x$, $0<y$, $x+y<case$の直線が、

正三角形の３辺に対応する直線に移る。残りの３辺も別の直線に移る。このような座標の線形変換を行うことで、

分割表が正三角形の内部の点として表現できるようになり、分割表同士の位置関係が可視化しやすくなる。

```r
#正三角形の内部に再配置
v1 <- c(1,0)
v2 <- c(1/2,sqrt(3)/2)
xy <- cbind(v1,v2) %*% rbind(cell_can_sub[,1],cell_can_sub[,2])
plot(t(xy),xlim=c(0,100),ylim=c(0,100))
```
